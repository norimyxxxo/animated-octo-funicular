```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('IWM.csv')
```

```python
date = data['Date'].values
price = data['Open'].values
```

```python
N = len(date)
tau = (N + 1) // 3
n = N - tau
X = np.array([price[i: n + i] for i in range(tau)])
C = (X @ X.T) / n
Lamda, V = np.linalg.eig(C)
Y = V.T @ X
```

```python
r = 100
V_r = V[:r]
Y_r = (V.T @ X)[:r,]
V_tau = V[-1, :r]
V_ast = V[:tau-1, :r]
Q = X[-tau+1 :]
```

```python
predicted_series = V_tau @ V_ast.T @ Q / (1 - V_tau @ V_tau.T)
predicted_series = np.append(price[:N-n], predicted_series)
```

```python
plt.figure(figsize=(20, 15))
plt.plot(price, 'b')
plt.plot(predicted_series, 'r')
plt.xlabel('Days')
plt.ylabel('price')
plt.show()
```
